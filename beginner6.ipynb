<h2>Matplotlib</h2>

<h3>Basic types of diagrams and methods of their construction</h3>

```python
import matplotlib.pyplot as plt
# plt is the commonly accepted alias for pyplot
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y)

# The simplest schedule
```

```python
plt.plot(x,y)

plt.savefig('sample_plot.png')
# Save image with diagram
```

```python
plt.plot(x,y)

plt.show()
# Display the diagram on the screen (not only on the laptop)
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]


plt.plot(x,y, label='y=2x', color='red', linewidth=2, marker='.', \
         linestyle='--', markersize=10, markeredgecolor='blue')

plt.show()
# Plot() function arguments

# To see all possible arguments and their values, use the documentation:
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
```

```python
plt.figure(figsize=(5,3), dpi=120)

plt.plot(x,y, label='y=2x', color='green', linewidth=2, marker='.', \
         linestyle='--', markersize=10, markeredgecolor='blue')

plt.legend()

plt.title('    Matplotlib')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4,])
plt.yticks([0,2,4,6,8,10])

plt.show()
# Layout of the diagram

# All available methods for pyplot design: https://matplotlib.org/stable/api/pyplot_summary.html
```

```python
import numpy as np

plt.plot(x,y, 'go--', label='y=2x', markeredgecolor='blue')

plt.legend()

plt.show()
# Using quick formatting
```

```python
import numpy as np

plt.plot(x,y, label='y=2x')

x2 = np.arange(0,4.5,0.5)
y2 = x2**2

plt.plot(x2[:6], y2[:6], label='y=X^2', color='green')

plt.plot(x2[5:], y2[5:], color='green', linestyle='--')

plt.legend()

plt.show()
# Building multiple shapes on one image
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

plt.show()
# Building a bar chart
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.barh(labels, values)

plt.show()
# Horizontal bar chart
```

```python
y = [35, 25, 25, 15]

plt.pie(y)

plt.show() 
# Pie chart
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)

plt.show()
# scatterplot
```

<h3>Using Matplotlib with Pandas</h3>

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Ali\Documents\python\gas_prices.csv")
df
```

```python
plt.plot(df['Year'], df['Canada'], 'b-', label='Canada')
plt.plot(df['Year'], df['France'], 'r-', label='France')

plt.legend()

plt.xlabel('Year')
plt.ylabel('Price, $')

plt.xticks(df['Year'][::3])

plt.show()
# Construction of diagrams from dataframes
```

```python
countries = ['France', 'Germany', 'Italy', 'UK']

for country in countries:
    plt.plot(df['Year'], df[country], label=country)

plt.legend()

plt.xlabel('Year')
plt.ylabel('Price, $')

plt.xticks(df['Year'][::3])
    
plt.show()
# Any number of lines on the graph
```

```python
dff = pd.read_csv(r"C:\Users\Ali\Documents\python\fifa_data.csv")

dff
```

```python
left = dff.loc[dff['Preferred Foot'] == 'Left'].count()[0]
right = dff.loc[dff['Preferred Foot'] == 'Right'].count()[0]

plt.pie([left, right], labels = [' ', ' '], autopct='%.0f%%')

plt.title('   ')

plt.show()
# Pie chart
# autopct='%.0f%%' is the format of the values of each of the sectors
```

```python
barcelona = dff.loc[dff['Club'] == "FC Barcelona"]['Overall']
madrid = dff.loc[dff['Club'] == "Real Madrid"]['Overall']
revs = dff.loc[dff['Club'] == "New England Revolution"]['Overall']

bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'])

plt.ylabel('FIFA Rating')

plt.show()
# Diagram of the type "Box plot" or "Box with whiskers"
```

```python
plt.hist(dff['Overall'])

plt.ylabel(' ')
plt.xlabel(' ')

plt.show()
# Building a histogram
```

<h2>Seaborn</h2>

```python
import seaborn as sns
# sns is a generally accepted nickname for seaborn
```

```python
tips = sns.load_dataset("tips")

tips
# We will use one of the datasets that come with Seaborn
```

```python
plt.style.use('seaborn-v0_8')

plt.scatter(tips['total_bill'], tips['tip'])

plt.show()
# Creating a dot plot in Matplotlib
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
# Creating a dot plot in Seaborn
# sns.set_theme() - to set the style of Seaborn renderings
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
# Dot chart "at maximums"
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip")
# Seaborn has built-in regression tools
```

```python
sns.displot(data=tips, x="total_bill")
# Histogram in Seaborn
```

```python
sns.displot(data=tips, x="total_bill")

plt.title("  ")

plt.xlabel(" ")

plt.ylabel(" ")
# Functions from Matplotlib also work for Seaborn
```

```python
sns.displot(data=tips, x="total_bill", col="time")
# Breakdown of histograms by field value
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
# Box Plot
```

```python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill")
# Visualization of the distribution in the form of a "violin"
```
