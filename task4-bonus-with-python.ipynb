```python
# Merge Facebook tables with campaign and adset info
facebook_ads_merged = (
    facebook_ads_basic_daily
    .merge(facebook_adset, on="adset_id", how="left")
    .merge(facebook_campaign, on="campaign_id", how="left")
)
```

```python
facebook_ads_merged.head()
```

```python
facebook_ads_merged.shape
```

```python
# Combine Facebook and Google ads
all_ads = pd.concat([
    facebook_ads_merged.assign(media_source="Facebook"),
    google_ads_basic_daily.assign(media_source="Google")
], ignore_index=True)
```

```python
all_ads.head()
```

```python
all_ads.shape
```

```python
# Calculate ROMI per campaign and adset
campaign_summary = (
    all_ads.groupby(["campaign_name", "media_source"], as_index=False)
    .agg({"spend": "sum", "value": "sum"})
)
campaign_summary["ROMI"] = ((campaign_summary["value"] - campaign_summary["spend"]) / campaign_summary["spend"]) * 100
```

```python
campaign_summary
```

```python
# Filter campaigns with total spend > 500000
campaign_summary_filtered = campaign_summary[campaign_summary["spend"] > 500000]
```

```python
campaign_summary_filtered
```

```python
# Find highest ROMI campaign
top_campaign = campaign_summary_filtered.loc[campaign_summary_filtered["ROMI"].idxmax()]
```

```python
top_campaign
```

```python
# Find top adset within that campaign (for Facebook only)
top_adset = (
    facebook_ads_merged[facebook_ads_merged["campaign_name"] == top_campaign["campaign_name"]]
    .groupby("adset_name", as_index=False)
    .agg({"spend": "sum", "value": "sum"})
)
top_adset["ROMI"] = ((top_adset["value"] - top_adset["spend"]) / top_adset["spend"]) * 100
top_adset_result = top_adset.loc[top_adset["ROMI"].idxmax()]

top_campaign, top_adset_result
```
