```python
import pandas as pd
```

```python
df_app=pd.read_csv(r"/Users/sametzubeyde/Downloads/applications(2.0).csv")
```

```python
df_app
```

```python
df_app = df_app.drop_duplicates(subset='applicant_id', keep='first')
```

```python
df_app
```

```python
df_app['External Rating']= df_app['External Rating'].fillna(0)
```

```python
df_app['Education level']= df_app['Education level'].fillna('Ortaokul')
```

```python
df_app
```

```python
df_industries=pd.read_csv(r"/Users/sametzubeyde/Downloads/industries(2.0).csv")
```

```python
df_industries
```

```python
df_industries = df_industries.rename(columns={'Score':'Industry Score'})
```

```python
df_industries.head()
```

```python
df_merged =pd.merge(df_app,df_industries)
```

```python
df_merged.head()
```

```python








df_merged[['applicant_id', 'Industry', 'Industry Score', 'FinalScore']].head()
```

```python
# Tarih sütununu datetime formatına dönüştür
df_merged['Applied at'] = pd.to_datetime(df_app['Applied at'], errors='coerce')
```

```python
# FinalScore hesaplama
df_merged['FinalScore'] = (
    ((df_merged['Age'].between(35, 55)) * 20) +                              # Yaş 35–55 arası
    ((df_merged['Applied at'].dt.weekday < 5) * 20) +                        # Hafta içi başvuru
    ((df_merged['Marital status'] == 'Married') * 20) +                      # Evli
    (df_merged['Location'].isin(['Kiev', 'Kiev çevresi']) * 10) +            # Kiev veya çevresi
    df_merged['Industry Score'] +                                            # Industry puanı
    ((df_merged['External Rating'] >= 7) * 20) -                             # External Rating ≥ 7
    ((df_merged['External Rating'] <= 2) * 20)                               # External Rating ≤ 2
)
```

```python
df_merged['FinalScore']
```

```python
# Amount eksikse veya External Rating = 0 ise FinalScore = 0
df_merged.loc[(df_merged['Amount'].isna()) | (df_merged['External Rating'] == 0), 'FinalScore'] = 0
```

```python
df_merged.head()
```

```python
# Puanı 0–100 arasında sınırla
df_merged['FinalScore'] = df_merged['FinalScore'].clip(0, 100)
```
